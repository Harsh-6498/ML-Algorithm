```python
import numpy as np
import pandas as pd 
```

```python
data=pd.read_csv("/content/train.csv")
```

```python
data.describe()
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
Age_mean=data['Age'].mean()
data['Age']=data ['Age'].fillna(Age_mean)
```

```python
data["Embarked"]=data["Embarked"].replace ('S',1)
```

```python
data["Embarked"]=data["Embarked"].replace ('C',0)

```

```python
data.head()
```

```python
data=data.drop(["Sex","Name","Embarked","Ticket","Cabin","PassengerId"],axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
y = data["Survived"]
data.drop(["Survived"], axis=1, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, y, random_state=42, test_size=0.2)

DCT = DecisionTreeClassifier(random_state=42)

DCT.fit(X_train, y_train)

y_pred = DCT.predict(X_test)

#print(y_pred)

accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix 
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RFC=RandomForestClassifier(n_estimators=500)
```

```python

RFC.fit(X_train, y_train)

y_pred = RFC.predict(X_test)

#print(y_pred)

accuracy_score(y_test, y_pred)
```
